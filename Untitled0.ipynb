```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('chinook.db')
```

```python
# از کاربر می‌پرسیم چند کلیدواژه وارد می‌خواد بده
keywords_input = input("چند کلیدواژه وارد کنید (مثال: Rock, Beatles, Love) — با کاما جدا کنید: ")
# تبدیل رشته به لیست کلیدواژه‌ها، حذف فاصله‌های اضافی
keywords = [kw.strip() for kw in keywords_input.split(',') if kw.strip()]
```

```python
sql_query = """
SELECT ...
WHERE (t.Name LIKE ? OR ar.Name LIKE ? OR g.Name LIKE ?)
  OR (t.Name LIKE ? OR ar.Name LIKE ? OR g.Name LIKE ?)
  OR (t.Name LIKE ? OR ar.Name LIKE ? OR g.Name LIKE ?)
ORDER BY t.Name;
"""


# اگر هیچ کلیدواژه‌ای وارد نشده باشه، می‌تونیم همه رکوردها رو بگیریم یا خطا بدیم
if not keywords:
    print("هیچ کلیدواژه‌ای وارد نشده است.")
    conn.close()
    exit()

# ساخت بخش WHERE دینامیک
where_clauses = []
params = []
for kw in keywords:
    clause = "(t.Name LIKE ? OR ar.Name LIKE ? OR g.Name LIKE ?)"
    where_clauses.append(clause)
    # برای هر کلیدواژه، ۳ پارامتر به لیست اضافه می‌کنیم
    like_param = f"%{kw}%"
    params.extend([like_param, like_param, like_param])

# ترکیب همه where_clauses با OR
where_sql = " OR ".join(where_clauses)

sql_query = f"""
SELECT
  t.TrackId,
  t.Name AS TrackName,
  al.Title AS AlbumTitle,
  ar.Name AS ArtistName,
  g.Name AS GenreName,
  t.UnitPrice
FROM tracks t
JOIN albums al ON t.AlbumId = al.AlbumId
JOIN artists ar ON al.ArtistId = ar.ArtistId
JOIN genres g ON t.GenreId = g.GenreId
WHERE {where_sql}
ORDER BY t.Name
"""

# اجرای کوئری
df = pd.read_sql_query(sql_query, conn, params=params)

if df.empty:
    print("هیچ نتیجه‌ای یافت نشد.")
else:
    # ممکنه رکوردها زیاد باشن: فقط ۲۰ تا اول
    display(df.head(20))
    # اگر بخوای همه رو ذخیره کنی به CSV:
    # df.to_csv('results.csv', index=False)
```
